### Visualizations

#### Univariate

```python
plt.hist(df['age'], bins=10, edgecolor='black', color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
df['sex'].value_counts().plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.show()
```

```python
df['smoker'].value_counts().plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Smoker Distribution')
plt.xlabel('Smoker')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['bmi'], bins=10, edgecolor='black', color='lightcoral')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.bar(df['children'].value_counts().index, df['children'].value_counts(), color='gold', edgecolor='black')
plt.title('Children Distribution')
plt.xlabel('Number of Children')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['charges'], bins=10, edgecolor='black', color='lightgreen')
plt.title('Charges Distribution')
plt.xlabel('Charges')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.bar(df['region'].value_counts().index, df['region'].value_counts(),color='green',edgecolor='black')
plt.title('Region distribution')
plt.xlabel('Region')
plt.ylabel('Frequency')
plt.show()
```

#### Bivariate

```python
sns.scatterplot(data=df,x=df.bmi,y=df.charges,hue=df.sex)
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='smoker', y='charges', data=df, palette='Set3')
plt.title('Box Plot for Smoker and Charges')
plt.xlabel('Smoker')
plt.ylabel('Charges')
plt.show()
```

```python
corr_matrix = df[['age', 'bmi', 'charges']].corr()

# Heatmap for correlation
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap for Correlation between Age, BMI, and Charges')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.swarmplot(x='children', y='charges', data=df, palette='Set2')
plt.title('Swarm Plot for Children and Charges')
plt.xlabel('Number of Children')
plt.ylabel('Charges')
plt.show()
```
